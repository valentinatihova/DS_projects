```python
def set_col_name(df, pivot = False):
    ''' Вспомогательная функция для лучшей читаемости названия полей в разрезе:
        - этап
        - тип_параметра
        - название_параметра
    '''
    df_pvt = pd.DataFrame([i.split('.') for i in df.columns.to_list()], 
                          columns = ['этап', 'тип_параметра', 'название_параметра'])
    if pivot:
        df_pvt = df_pvt.groupby(['этап', 'тип_параметра'])['название_параметра'].agg(['unique']).reset_index()
    return df_pvt
```

### Проверка расчётов эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака `rougher.output.recovery`. 

Эффективность обогощения рассчитывается по формуле:
$$Recovery = {{C\times(F-T)} \over {F\times(C-T)}}\times 100\%,$$
где:
* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

```python
def rough_recov_calc(df):
    C = df['rougher.output.concentrate_au']
    F = df['rougher.input.feed_au']
    T = df['rougher.output.tail_au']

    rough_recov_calc = C * (F - T) / (F * (C - T)) * 100
    
    return rough_recov_calc
```

```python
df_train['rough_recov_calc'] = rough_recov_calc(df_train)
```

Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.
Метрика sMAPE вычисляется так:
$$sMAPE = {1\over N} \sum_{i=1}^N {|y_i - \widehat{y_i}| \over ((|y_i|+|\widehat{y_i}|)/2)} \times 100\%   $$


Найдем `sMAPE` между вашими расчётами и значением признака

```python
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))
```

```python
smape = smape(df_train['rougher.output.recovery'], df_train['rough_recov_calc'])
print('sMAPE между расчётами и значением признака:', smape)
```

Показатель sMAPE очень мал, получается, что наши расчеты верны.

Для анализа конфентрации металлов посмотрим графики завесимости между концентрацией по времени для каждого этапа.
Для отрисовки графика выделим метрики, которые отвечают за концентрацию.

```python
col_concentrate = df_train.filter(like='concentrate').columns.tolist()
list_to_remove = ['final.output.concentrate_sol','primary_cleaner.output.concentrate_sol',
                  'rougher.calculation.sulfate_to_au_concentrate','rougher.output.concentrate_sol']
col_concentrate = list(set(col_concentrate) - set(list_to_remove))
```

```python
def daily_plot(me, step):
    ''' Функция отрисовки временного графика в разрезе:
        - металлов me
        - этапам step
    '''
    sns.set()
    f, axes = plt.subplots(1, 1, figsize=(16, 6))
    
    axes.set_ylabel('концентрация', fontsize=15)
    
    df_train['date'] = pd.to_datetime(df_train['date'])
    df_train_daily = df_train.resample('D', on='date').mean().reset_index()
    df_train_daily = df_train_daily.set_index('date')
    df_train_daily.index.name = ''
    
    if me:
        cols_me = list(filter(lambda k: me in k, col_concentrate))
        axes.set_title("Концентрация " + me, fontsize=18)
        sns.lineplot(data=df_train_daily[cols_me]);
    
    if step:
        cols_step = list(filter(lambda k: step in k, col_concentrate))
        axes.set_title("Концентрация " + step, fontsize=18)
        sns.lineplot(data=df_train_daily[cols_step]);
```

```python
for me in ('au', 'ag', 'pb'):
    daily_plot(me, '')
```

В разрезе металлов цепочка **флотация => первичная очистка => финал**:
* концентрация золота `au` увеличивается, сереба `ag` уменьшается, свинца `pb`- уведичивается после флотации

```python
for step in ('rougher', 'primary_cleaner', 'final'):
    daily_plot(step, '')
```

В разрезе этапов цепочка **флотация => первичная очистка => финал**:
* концентрация золота увеличивается с каждым этапом, а еще она выше по сравнению с другими металлами

### Суммарная концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах

## Модель

### Итоговая sMAPE

Ранее уже была написана функция sMAPE, напишем итоговую функцию для метрики качества.

```python
def final_smape(smape_A, smape_B):
    return 0.25 * smape_A + 0.75 * smape_B
```

### Обучение модели и проверка качества кросс-валидацией

Cделаем одинаковые признаки в обучающей и тестовой выборках.

```python
col_diff = set(df_train_new.columns) - set(df_test.columns)

df_train_new = df_train_new.drop(col_diff, axis = 1)

print('Итоговые признаки для модели (train & test):')
set_col_name(df_train_new, True)
```

Выделим признаки:

```python
df_test1= df_test.set_index('date', inplace=False, drop=True)
df_train_new1 = df_train_new.set_index('date', inplace=False, drop=True)
df_full1 = df_full.set_index('date', inplace=False, drop=True)
```

```python
X_train_rougher = df_train_new1.drop(['rougher.output.recovery', 'final.output.recovery'], axis = 1)
X_test_rougher = df_test1.drop(['rougher.output.recovery', 'final.output.recovery'], axis = 1)

X_train_final = df_train_new1.drop(['rougher.output.recovery', 'final.output.recovery'], axis = 1)
X_test_final = df_test1.drop(['rougher.output.recovery', 'final.output.recovery'], axis = 1)
```

и цель, используя индекс по дате:

```python
def set_target(x, col):
    ''' Выделаю только нужные строки по индексу'''
    return df_full1.query('index in @x.index')[col]
```

```python
for i in (df_full1, df_test1, df_train_new1):
    print(F'index size: min = {i.index.min()}, max = {i.index.max()}')
```

```python
def dummies_index(df1, col):
    ''' Через мердж обучающей\тестовой выборок и полной получаю нужный столбец'''
    megre_df = df1.merge(df_full,
                             left_index = True,
                             right_index = True,
                             how='left',
                             suffixes=('', 'del'))

    return megre_df[col]
```

```python
y_train_rougher = dummies_index(df_train_new1, 'rougher.output.recovery')
y_test_rougher = dummies_index(df_test1, 'rougher.output.recovery')
y_train_final = dummies_index(df_train_new1, 'final.output.recovery')
y_test_final = dummies_index(df_test1, 'final.output.recovery')
```

```python
print('Сравнение признаков для модели на этапе флотации:', set(X_train_rougher) == set(X_test_rougher))
print('Сравнение признаков для модели на этапе финал:', set(X_train_final) == set(X_test_final))
```

Проверим размерность выборок перед началом обучения модели.

```python
print('Признаки в обучающей выборке:         ', X_train_rougher.shape, X_train_final.shape)
print('Признаки в тестовой выборке:          ', X_test_rougher.shape, X_test_final.shape)
print('Целевые признаки в обучающей выборке: ', y_train_rougher.shape, y_train_final.shape)
print('Целевые признаки в тестовой выборке:  ', y_test_rougher.shape, y_test_final.shape)
```

Введем счетчик из метики производительности или потери функции make_scorer.

```python
def smape(y, y_pred):
    return 1 / len(y) * sum(abs(y - y_pred) / ( (abs(y) + abs(y_pred)) * 0.5) ) * 100
```

```python
smape_score = make_scorer(smape)
random_state = 12345
```

Напишем вспомогательную функцию для вывода результатов.

```python
def print_res(a,b,c=True):
    print(f'Флотация: {a:,.4}')
    print(f'Финал: {b:,.4}')
    if c:
        print(f'final_smape: {c:,.4}')
```

#### Линейная регрессия 

```python
def model_LR(X, y):

    model_LR = LinearRegression().fit(X, y)
    smape_LR = cross_val_score(model_LR, 
                               X, 
                               y, 
                               scoring=smape_score,
                               cv = 5)
    return abs(smape_LR.mean())
```

```python
model_LR_rougher = model_LR(X_train_rougher, y_train_rougher)
model_LR_final = model_LR(X_train_final, y_train_final)
final_smape_LR = final_smape(model_LR_rougher, model_LR_final)
```

#### Дерево решений

```python
def model_DT(X, y):
    
    params = {'min_samples_split': range(2, 10, 2), 'max_depth': range(4,8,2)}
    model_DT = GridSearchCV(DecisionTreeRegressor(random_state=random_state),
                            param_grid=params,
                            scoring=smape_score, 
                            cv=5, 
                            refit=True)

    best_params = model_DT.fit(X, y).best_params_
    
    model_DT = DecisionTreeRegressor(**best_params).fit(X, y)
    
    smape_DT = cross_val_score(model_DT, 
                               X,
                               y, 
                               scoring=smape_score, 
                               cv=5)
    return abs(smape_DT.mean())   
```

```python
model_DT_rougher = model_DT(X_train_rougher, y_train_rougher)
model_DT_final = model_DT(X_train_final, y_train_final)
final_smape_DT = final_smape(model_DT_rougher, model_DT_final)
```

#### Проверка модели на адекватность

Нужно проверить, насколько качественно предсказывает наша модель. Используем для этого метод DummyRegression:

```python
def model_DM(X_train, X_test, y_train, y_test):
    
    model_DM = DummyRegressor(strategy='mean').fit(X_train, y_train)
    y_pred = model_DM.predict(X_test)
    
    smape_DM = smape(y_test, y_pred)
    return smape_DM
```

```python
model_DM_rougher = model_DM(X_train_rougher, X_test_rougher, y_train_rougher, y_test_rougher)
model_DM_final = model_DM(X_train_final, X_test_final, y_train_final, y_test_final)
final_smape_DM = final_smape(model_DM_rougher, model_DM_final)
```

Посмотрим на итоги:

```python
print('===== Линейная регрессия:')
print_res(model_LR_rougher, model_LR_final, final_smape_LR)
print()
print('===== Дерево решений:')
print_res(model_DT_rougher, model_DT_final, final_smape_DT)
print()
print('===== Константа:')
print_res(model_DM_rougher, model_DM_final, final_smape_DM)
```

Итого, лучшая модель с точки зрения качества - дерево решений. Проверим ее на тестовой выборке.
